## 0. Прототип модели
Мы используем нейросеть с одним скрытым слоем.


```python
def sigmoid(z):
    return 1/(1 + np.exp(-1 * z))
```

```python
l_in = 400 # amount of features
l_1 = 25 # different studies
l_out = 10 # number of outputs
k = l_out
```

Также в этой модели мы хотим повернуть матрицы X и Y, и применить OHE для Y, так как наша стоимостная функция подразумевает только бинарный выход

```python
def one_hot_encoding(y, k = 10):
    m,n = y.shape
    
    Y = np.zeros((m,k))
    for i in range(len(y.reshape(m))):
        Y[i,y[i] - 1] = 1
    return Y

X = X.T # must be 400 x 5000
Y = one_hot_encoding(Y).T # must be 10 x 5000
print("X in", X.shape,"\nY in", Y.shape)
```

## 1. Feedforward Propagation
Сначала проходим всю модель и находим стоимостную функцию.

Для начала нам понадобится две функции: разворачивание и сворачивание весов в вектор. Чтобы дальше было легче реализуем функции сейчас

```python
def weights_wrap(theta1, theta2):
    return np.concatenate((theta1.flatten(),theta2.flatten()))

def weights_unwrap(theta, l_in,l_1,l_out):
    new_weight_1 = theta[0:l_1 * (l_in +1)].reshape(l_1, l_in + 1)
    new_weight_2 = theta[l_1 * (l_in +1):].reshape(l_out, l_1 + 1)
    return new_weight_1, new_weight_2
```

Также полезной функцией будет добавление bias unit к производной матрице:

```python
def add_bias(matrix, direction = 0):
    m,n = matrix.shape
    
    shape = (m,1)
    if direction == 0:
        shape = (1,n)

    return np.concatenate((np.ones(shape),matrix),direction) 
```

Помимо прочего, сразу добавим возможность регулирования функции

```python
def cost(X,y,theta, l_in,l_1,l_out, reg = 0):
    Theta1, Theta2 = weights_unwrap(theta,l_in,l_1,l_out)
    
    a1 = add_bias(X) # (l_in + 1) x m
    z2 = Theta1 @ a1 # l_1 x m
    a2 = add_bias(sigmoid(z2)) # (l_1 + 1) x m
    z3 = Theta2 @ a2 # k x m
    a3 = sigmoid(z3) # k x m
    h = a3
    
    J = np.sum(-y * np.log(h) - (1-y) * np.log(1 - h))/m
    
    weight_minimizer = 0
    
    if reg != 0:
        weight_minimizer += np.sum(Theta1[:,1:]**2)
        weight_minimizer += np.sum(Theta2[:,1:]**2)
    
    J += reg / (2*m) * weight_minimizer
    
    return J
```

## 2. Backpropagation
Теперь мы хотим найти ошибки наших нейронов. Сперва сделаем несколько вспомогательных функций: производную от сигмоидной, случайную инициализацию и проверку градиента:

```python
def sigmoid_grad(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

У случайной инициализации весов есть предпочтительная формула создания окрестности $ \epsilon = \frac{\sqrt{6}}{\sqrt{l_{in} + l_{out}}} $, однако в обучающих материалах нам рекомендовано взять постоянную $ \epsilon = 0.12 $ — так и поступим.

```python
def random_weight(l_in, l_out):
    epsilon_init = 0.12
    return (np.random.rand(l_out, l_in) * 2 * epsilon_init - epsilon_init).T
```

```python
def grad_check(J, theta):
    num_grad = np.zeros(theta.shape)
    perturb = np.zeros(theta.shape)
    e = 1e-4
    
    for i in range(theta.shape[0]):
        perturb[i] = e
        loss1 = J(theta - perturb)
        loss2 = J(theta + perturb)
        num_grad[i] = (loss2 - loss1) / (2 * e)
        
        perturb[i] = 0
    return num_grad
```

```python
def full_cost(X,y,theta, l_in,l_1,l_out, reg = 0):
    Thet_1, Thet_2 = weights_unwrap(theta,l_in,l_1,l_out)
    
    # Feedforward
    a1 = add_bias(X) # (l_in + 1) x m
    z2 = Thet_1 @ a1 # l_1 x m
    a2 = add_bias(sigmoid(z2)) # (l_1 + 1) x m
    z3 = Thet_2 @ a2 # k x m
    a3 = sigmoid(z3) # k x m
    h = a3
      
    J = np.sum(-y * np.log(h) - (1-y) * np.log(1 - h))/m
    
    weight_minimizer = 0
    
    if reg != 0:
        weight_minimizer += np.sum(Theta1[:,1:]**2)
        weight_minimizer += np.sum(Theta2[:,1:]**2)
    
    J += reg / (2*m) * weight_minimizer
    
    # Backpropagation
    d3 = a3 - y # k x m
    d2 = Thet_2.T @ d3 * add_bias(sigmoid_grad(z2)) # (l_1 + 1) x m
    
    D2 = (d3 @ a2.T)/ m # k x (l_1 + 1)
    D1 = (d2[1:,:] @ a1.T)/ m # l_1 x (l_in + 1)
    
    grad = weights_wrap(D1,D2)
    return J, grad
```

```python
def cost_wrap(X,Y,theta):
    return full_cost(X,Y,theta, l_in,l_1,l_out)

def gradient_descent(X, y, theta, alpha, num_iters, regulator = None):
    m = X.shape[1]
    J = 0
    J_history = np.zeros((num_iters,1))
    theta_history = np.zeros((num_iters, theta.shape[0]))
    
    for i in range(num_iters):
        J, delta = cost_wrap(X,y,theta)
        J_history[i] = J
        
        theta = theta - alpha * delta
        theta_history[i] = theta
        
    return (J, J_history,theta, theta_history)
```

```python
test_theta_1 = random_weight(25,401)
test_theta_2 = random_weight(10,26)
weights2 = weights_wrap(test_theta_1,test_theta_2)


J, dJ = full_cost(X,Y,weights2, l_in,l_1,l_out)
print("Original J", J)

new_J, history_J, weights3, history = gradient_descent(X,Y, weights2, 2, 1000)
print("new J", new_J)
print(100 - (new_J/J* 100) ,"%")

iterations = np.arange(history.shape[0])

plt.plot(iterations, history_J)
plt.show()

plt.plot(iterations,history)
plt.xlabel("Iterations")
plt.ylabel("params")
plt.title(label="Yo")
plt.show()
```

```python
def predict(X):
    Thet_1, Thet_2 = weights_unwrap(weights3,l_in,l_1,l_out)
    
    a1 = add_bias(X)
    z2 = Thet_1 @ a1
    a2 = add_bias(sigmoid(z2))
    z3 = Thet_2 @ a2
    h = sigmoid(z3)
    
    return np.where(h == np.amax(h))[0][0]

predict(X[:,4000].reshape(400,1))
```
