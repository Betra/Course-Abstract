```python
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
import utils
from scipy.ndimage.filters import gaussian_filter1d


%cd ~/Documents/ML/machine-learning-ex5/ex5/

mat = loadmat('ex5data1.mat')
X = np.array(mat['X'])
y = np.array(mat['y'])

m, n = X.shape
```

```python
def hypothesis(X, theta):
    return X @ theta

def linear_cost(X, y, theta, regularization = 0):
    m, n = X.shape
    h = hypothesis(X, theta).reshape((m,1)) # m x 1
    
    reg_theta = np.sum(theta[1:] ** 2) # theta must be one-dimensional
    
    J = 1/(2*m) * np.sum((h - y) ** 2) + regularization/(2*m) * reg_theta
    
    delta = (1/m * (h-y).T @ X + regularization/m * theta).reshape(n) # one-dimensional
    
#     print(J, delta)
    return J, delta
```

```python
def add_bias(matrix, direction = 0):
    m,n = matrix.shape
    
    shape = (m,1)
    if direction == 0:
        shape = (1,n)

    return np.concatenate((np.ones(shape),matrix),direction) 

def normal_equation(X,y):
    theta = np.linalg.pinv(X.T @ X) @ X.T @ y
    return theta

def train_model(X, y, 
                graphics = False):
    X_local = add_bias(X,1)
    m, n = X_local.shape
    theta = np.zeros(n)
    
    LEARNING_RATE = 0.00231
    REGULARIZATION = 0
    ITERATIONS = 2000
    
    # Initial values
    J, delta = linear_cost(X_local, y,  theta, 1)
    

    # We can't use gradient descent because it requires
    # learning rate to differ on different datasets
    # instead we are going to use normal equation
    
#     new_theta, J, J_history,theta_history = utils.optimize_gradient(
#         X_local,y, theta, linear_cost,LEARNING_RATE , 
#         regularization = REGULARIZATION, max_iterations=ITERATIONS)
    
    new_theta = normal_equation(X_local,y).reshape(n)
    [J, delta_normal] = linear_cost(X_local,y,new_theta,1)
    
    
    if graphics is not True:
        return J, new_theta
    
    print("Final J = %.3f" %J)
    
    # Drawing
    plt.plot(X_local ,y,"bx", label = "Training examples")
    plt.plot(X_local, 
             hypothesis(X_local,new_theta), 
             "r", label = "Hypothesis")
    
    plt.title(label = "Dataset representation with hypothesis")
    plt.legend()
    plt.show()

train_model(X, y, graphics = True)
```

```python
def build_learning_curve(Xt):
    m, n = Xt.shape
    
    slice_size = np.arange(1,m + 1)
    Jt_history = np.zeros(m)
    Jcv_history = np.zeros(m)
    
    for size in slice_size:
        X_sliced = Xt[:size,:]
        y_sliced = y[:size]
        
        [Jt, theta_local] = train_model(X_sliced,y_sliced)
        [Jcv, delta] = linear_cost(add_bias(Xt,1),y, theta_local, 10)
        
        Jt_history[size-1] = Jt
        Jcv_history[size-1] = Jcv
    
    # Drawing
    plt.plot(slice_size,Jt_history, label="Training")
    plt.plot(slice_size,Jcv_history, label="Cross Validation")
    plt.xlabel("Dataset size")
    plt.ylabel("Cost function")
    
    plt.legend()
    plt.show()
    
build_learning_curve(X)
```

```python
def poly_features(X,p):
    m, n = X.shape
    polyX = np.zeros((m,p))
    
    for i in range(p):
        polyX[:,i] = (X ** (i+1)).reshape(m)
    return polyX
```

```python
def polylinear_cost(X, y, theta, regularization = 0, poly = 8):
    m, n = X.shape
    
    h = hypothesis(X, theta).reshape((m,1)) # m x 1
    
    reg_theta = np.sum(theta[1:] ** 2) # theta must be one-dimensional
    
    J = 1/(2*m) * np.sum((h - y) ** 2) + regularization/(2*m) * reg_theta
    
    delta = (1/m * (h-y).T @ X + regularization/m * theta).reshape(n) # one-dimensional
    
    return J, delta
```
